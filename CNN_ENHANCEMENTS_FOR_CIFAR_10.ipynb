```python
# === Imports ===
import os  # For system operations (setting environment variables)
import numpy as np  # For numerical operations
import tensorflow as tf  # For deep learning
from tensorflow.keras import layers, models, backend as K  # Keras components
from tensorflow.keras.datasets import cifar10  # CIFAR-10 dataset
from tensorflow.keras.utils import to_categorical  # One-hot encoding
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau  # Training callbacks
import matplotlib.pyplot as plt  # For plotting
import albumentations as A  # For data augmentation
import cv2, gc  # OpenCV and garbage collection
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay  # Confusion metrics
from sklearn.metrics import classification_report


# Suppress TensorFlow logs (optional)
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

# === Global Training Settings ===
EPOCHS = 50  # Number of training epochs
BATCH_SIZE = 128  # Number of samples processed per batch - depends on the computing power

# === Load and Preprocess CIFAR-10 Dataset ===
(x_train, y_train), (x_test, y_test) = cifar10.load_data()  # Load data
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)

x_train, x_test = x_train.astype("bfloat16") / 255.0, x_test.astype("bfloat16") / 255.0  # Normalize to [0, 1]
y_train_cat, y_test_cat = to_categorical(y_train, 10), to_categorical(y_test, 10)  # One-hot encode labels

# === Visualization of Distribution and Image Samples ===
# Get the unique classes and their counts in the training set
unique_classes, class_counts = np.unique(y_train, return_counts=True)

# Create a bar chart of the class distribution
plt.figure(figsize=(8, 6))
plt.bar(unique_classes, class_counts)
plt.title('Class Distribution in CIFAR-10 Training Set')
plt.xlabel('Class Label')
plt.ylabel('Number of Samples')
plt.xticks(unique_classes)  # Set x-axis ticks to class labels
plt.show()

# Define class names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Display 6 random images from the training set
plt.figure(figsize=(10, 10))
for i in range(6):
    random_index = np.random.randint(0, len(x_train))
    image = x_train[random_index]
    label = y_train[random_index][0]  # Get the label from the array
    plt.subplot(3, 3, i + 1)
    plt.imshow(image)
    plt.title(f"Label: {class_names[label]}")
    plt.axis('off')
plt.show()

# === Data Augmentation with Albumentations ===
albu_transform = A.Compose([  # Define augmentation pipeline
    A.HorizontalFlip(p=0.5),  # Random horizontal flips
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=15, p=0.5),  # Random shifts, scales, and rotations
    A.RandomBrightnessContrast(p=0.2),  # Random brightness and contrast adjustments
    A.HueSaturationValue(p=0.2),  # Random hue, saturation, and value adjustments
])

# === Data Generator with Augmentations ===
def albumentations_generator(x, y, batch_size): # x – Input Images, y – Labels
    # Generates batches of augmented data.
    while True:
        indices = np.random.permutation(len(x))  # Shuffle data indices
        for i in range(0, len(x), batch_size):  # Iterate over batches
            batch_x, batch_y = [], y[indices[i:i+batch_size]]  # Initialize batch data: batch_x contains augmented images; batch_y holds corresponding labels
            for j in indices[i:i+batch_size]:  # Process each sample in the batch
                image = albu_transform(image=(x[j] * 255).astype(np.uint8))['image']  # Apply augmentations - albu_transform(image=image_array)['image'], x[j] in [0.0, 1.0], uint8 = 8 bit (255)
                batch_x.append(image.astype(np.float32) / 255.0)  # Normalize and add to batch
            yield np.stack(batch_x), batch_y  # Yield the batch


# Sample Image After Augmentation
idx = np.random.randint(len(x_train))
sample_image = x_train[idx]
augmented_image1 = albu_transform(image=(sample_image * 255).astype(np.uint8))['image']
augmented_image2 = albu_transform(image=(sample_image * 255).astype(np.uint8))['image']

# Normalize augmented image
augmented_image1 = augmented_image1.astype(np.float32) / 255.0
augmented_image2 = augmented_image2.astype(np.float32) / 255.0

# Display the original and augmented images
plt.figure(figsize=(9, 3))
images = [("Original", sample_image), ("Augmented 1", augmented_image1), ("Augmented 2", augmented_image2)]

for i, (title, img) in enumerate(images, 1):
    plt.subplot(1, 3, i)
    plt.title(title)
    plt.imshow(img)
    plt.axis('off')

plt.tight_layout()
plt.show()

# === Basic CNN model with optional residual blocks, SE blocks, and vision transformer ===
def build_cnn(input_shape=(32, 32, 3), num_classes=10, use_residual=False, use_se=False, use_transformer=False, return_features=False): # 32×32 with 3 RGB channels
    # Squeeze-and-Excitation block
    def se_block(x):
        se = layers.GlobalAveragePooling2D()(x) # Compresses each channel to a single scalar by average
        se = layers.Dense(x.shape[-1] // 16, activation='relu')(se) # x.shape[-1] = channel, r = reduction ratio, typically 16, relu in hidden layer
        se = layers.Dense(x.shape[-1], activation='sigmoid')(se) # sigmoid in output layer
        return layers.Multiply()([x, layers.Reshape((1, 1, x.shape[-1]))(se)]) # scales each channel in the image features (x) by a learned weight from se

    # Residual block with optional SE block
    def residual_block(x, filters):
        shortcut = x  # Input
        # Conv2D -> BatchNorm -> ReLU -> Conv2D -> BatchNorm
        x = layers.Conv2D(filters, 3, padding='same')(x)  # Applies x filters, each with a 3×3 kernel, padding = Keeps output size same
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)
        x = layers.Conv2D(filters, 3, padding='same')(x)
        x = layers.BatchNormalization()(x)
        if use_se:
            x = se_block(x)
        if shortcut.shape[-1] != filters:
            shortcut = layers.Conv2D(filters, 1, padding='same')(shortcut)  # Now both x and shortcut have x channels -> can add together
            shortcut = layers.BatchNormalization()(shortcut)
        return layers.ReLU()(layers.Add()([x, shortcut])) # Adds the shortcut to the main output and applies ReLU to the result — a core idea in ResNet.

    inputs = layers.Input(shape=input_shape)
    x = layers.Conv2D(64, 3, padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)

    # CNN backbone
    for filters in [64, 128, 256]:
        if use_residual:
            x = residual_block(x, filters)
        else:
            x = layers.Conv2D(filters, 3, padding='same', activation='relu')(x)
            x = layers.BatchNormalization()(x)

        if use_se:
            x = se_block(x)

        x = layers.MaxPooling2D()(x)
        x = layers.Dropout(0.2)(x)  # reduce neuron -> for generalization

    # Return_features=True makes the model output feature maps(excluding global average pooling, dense layers, and final classification) which RDNet uses for further processing
    if return_features:
        return models.Model(inputs, x, name="Backbone")

    # Vision transformer block
    if use_transformer:
        x = layers.Reshape((-1, x.shape[-1]))(x)  # Patch Embedding(Flatten image into patch tokens): (50000, 32, 32, 3) to (50000, 32*32 = 1024, 3)
        attn = layers.MultiHeadAttention(num_heads=4, key_dim=64)(x, x) # Transformer Encoder(Multi-Head Self-Attention block): 4 parallel heads, 64 dimensions
        x = layers.GlobalAveragePooling1D()(attn)
        # MLP Head(Dense layers for final classification): reuse CNN structure in next section, Dense(512) → Dense(10, Softmax)
    else:
        x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.2)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)  # dense to 10 class
    return models.Model(inputs, outputs)

# === RDNet Model Definition ===
# Feature Mixer Block for RDNet
# Following a DenseNet-style pattern: Batch Normalization -> ReLU -> Convolution -> Dropout
class FeatureMixerBlock(layers.Layer):
    def __init__(self, growth_rate):  # init: Setup
        super().__init__()  # call the parent class
        self.bn = layers.BatchNormalization()
        self.conv = layers.Conv2D(growth_rate, 3, padding='same')
        self.drop = layers.Dropout(0.2)

    def call(self, x, training=False):  # call: Forward Pass
        out = self.conv(tf.nn.relu(self.bn(x, training=training)))  # tf.nn = TensorFlow neural network operations -> relu activation
        out = self.drop(out, training=training)
        return tf.concat([x, out], axis=-1) # Concatenate original input (x) and new output -> result, axis = -1 -> last dimension

# Transition Layer for RDNet - downsizing
class TransitionLayer(layers.Layer):
    # Reduces the number of channels using 1×1 Conv2D
    def __init__(self, out_channels):
        super().__init__()
        self.bn = layers.BatchNormalization()
        self.conv = layers.Conv2D(out_channels, 1)
        self.drop = layers.Dropout(0.2)

    def call(self, x, training=False):
        x = self.bn(x, training=training)
        x = self.conv(x)
        x = self.drop(x, training=training)
        return x

# Full RDNet Model
class RDNet(tf.keras.Model):
    def __init__(self, backbone_model, growth_rate=12, block_config=(6, 6, 6)): # block_config = number of blocks in each stage, growth_rate = number of new feature maps produced by each layer
        super().__init__()
        self.backbone = backbone_model  # Use CNN as a core
        self.blocks = []  # empty list to store layers
        channels = self.backbone.output_shape[-1] # use previous channels
        for i, num_blocks in enumerate(block_config):
            for _ in range(num_blocks): # Feature Mixer Blocks
                self.blocks.append(FeatureMixerBlock(growth_rate))
                channels += growth_rate
            if i < len(block_config) - 1: # Transition Layers, len((6, 6, 6)) = 3
                out_ch = channels // 2  # /2 and rounding down
                self.blocks.append(TransitionLayer(out_ch))
                channels = out_ch
        self.pool = layers.GlobalAveragePooling2D()
        self.fc = layers.Dense(10)

    def call(self, x, training=False):
        x = self.backbone(x, training=training)
        for layer in self.blocks:
            x = layer(x, training=training)
        return self.fc(self.pool(x))

# === Callbacks ===
early_stop = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True) # If validation accuracy doesn’t improve for 10 consecutive epochs, training stops, and reverts to the best checkpoint
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5) # If validation loss doesn't improve for 5 epochs, reduces learning rate /= 0.5

# === Training & Evaluation Helper ===
def run_and_evaluate(model, name, train_data, val_data, epochs=EPOCHS, steps_per_epoch=None):
    # Compiles, trains, and evaluates a model
    # Use Adam optimizer with a learning rate of 0.0005 (0.001 is default, 0.0005 is more stable, 0.0001 is for fine-tuning or big models)
    #a Adm is a popular and efficient choice for training deep networks
    model.compile(optimizer=tf.keras.optimizers.Adam(0.0005),
                  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True) if 'RDNet' in name else 'categorical_crossentropy',
                  metrics=['accuracy'])
    # If RDNet -> use 'oricasparse_categl_crossentropy' (for integer labels), else -> use 'categorical_crossentropy' (for one-hot labels)
    # last layer does not include softmax -> from_logits=True

    with tf.device('/GPU:0'): # Use GPU if available
        if 'RDNet' in name:
            history = model.fit(train_data,
                      validation_data=val_data,
                      epochs=epochs,
                      callbacks=[early_stop, reduce_lr],
                      verbose=1)
            _, acc = model.evaluate(val_data, verbose=0)

        # Check if train_data is albumentations_generator
        elif 'Augmentation' in name:
            history = model.fit(train_data,
                      steps_per_epoch=steps_per_epoch,
                      validation_data=val_data,
                      epochs=epochs,
                      callbacks=[early_stop, reduce_lr],
                      verbose=1)
            _, acc = model.evaluate(val_data[0], val_data[1], verbose=0)
        else:
            x_train, y_train = train_data
            x_val, y_val = val_data
            history = model.fit(x_train,
                      y_train,
                      validation_data=(x_val, y_val),
                      batch_size=BATCH_SIZE,
                      epochs=epochs,
                      callbacks=[early_stop, reduce_lr],
                      verbose=1)
            _, acc = model.evaluate(x_val, y_val, verbose=0)  # [loss, accuracy], "_" = ignore loss

        # Plot Train vs Validation Accuracy for Each Model
        plt.figure(figsize=(8, 4))
        plt.plot(history.history['accuracy'], label='Train Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title(f'{name} - Accuracy per Epoch')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        # After model evaluation
        if name == 'CNN + RDNet':
            y_pred = model.predict(val_ds)
            y_pred_labels = np.argmax(y_pred, axis=1)
            y_true = np.concatenate([y for _, y in val_ds], axis=0)
        else:
            y_pred = model.predict(x_test)
            y_pred_labels = np.argmax(y_pred, axis=1)
            y_true = y_test.flatten()  # original labels are shape (10000, 1)

        def plot_confusion_matrix(y_true, y_pred, class_names, title):
            cm = confusion_matrix(y_true, y_pred)
            disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
            disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
            plt.title(title)
            plt.tight_layout()
            plt.show()


        # Plot confusion matrix
        plot_confusion_matrix(y_true, y_pred_labels, class_names=class_names, title=f"{name} - Confusion Matrix")

        # classification_report
        print(classification_report(y_true, y_pred_labels, target_names=class_names))


    return acc, history

# === Main Evaluation Script ===
# initializes tracking variables
results = {}
histories = {}

best_acc = 0
best_model = None
best_model_name = ''

configs = {
    'Basic CNN': dict(use_residual=False, use_se=False, use_transformer=False),
    'CNN + Augmentation': dict(use_residual=False, use_se=False, use_transformer=False),
    'CNN + Residual': dict(use_residual=True, use_se=False, use_transformer=False),
    'CNN + SE': dict(use_residual=True, use_se=True, use_transformer=False),
    'CNN + Transformer': dict(use_residual=False, use_se=False, use_transformer=True),
    'CNN + RDNet': dict(use_residual=False, use_se=False, use_transformer=False, RDNet=True),
    'CNN + Augmentation + RES + SE + TF': dict(use_residual=True, use_se=True, use_transformer=True),
    'CNN + Augmentation + RES + SE + RDNet': dict(use_residual=True, use_se=True, RDNet=True)
}

for name, cfg in configs.items():
    print(name)
    if 'RDNet' in name:
        with tf.device('/GPU:0'):  # Use GPU if available
            train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).shuffle(5000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
            val_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test)).batch(BATCH_SIZE)
            # from_tensor_slices: Converts x_train and y_train into a Dataset
            # shuffle(5000): Randomizes the data
            # batch(BATCH_SIZE): Divides data into batches
            # prefetch(tf.data.AUTOTUNE): Preloads the next batch during training to speed up processing.

        # Builds RDNet based on CNN architecture
        backbone = build_cnn(use_residual=False, use_se=False, use_transformer=False, return_features=True)
        model = RDNet(backbone_model=backbone)

        # Build the model based on the input shape
        _ = model(tf.ones((1, *x_train.shape[1:])))
        # x_train.shape[1:] -> (32, 32, 3)
        # tf.ones((1, 32, 32, 3)) → Creates a batch of 1 image, filled with all ones

        model.summary()

        acc, _ = run_and_evaluate(model, name, train_ds, val_ds)
        results[name] = acc

    else:
        model = build_cnn(**cfg)
        model.summary()
        if 'Augmentation' in name:
            train_gen = albumentations_generator(x_train, y_train_cat, BATCH_SIZE)
            steps = len(x_train) // BATCH_SIZE
            acc, _ = run_and_evaluate(model, name, train_data=train_gen, val_data=(x_test, y_test_cat), steps_per_epoch=steps)
        else:
            acc, _ = run_and_evaluate(model, name, train_data=(x_train, y_train_cat), val_data=(x_test, y_test_cat))
        results[name] = acc

    # Save if best model
    if acc > best_acc:
        best_acc = acc
        best_model = model
        best_model_name = name


    # Clear session and collect garbage to prevent out of memory
    K.clear_session(); gc.collect()

# === Plot training output ===
plt.figure(figsize=(10, 5))
plt.bar(results.keys(), [v[0] * 100 if isinstance(v, tuple) else v * 100 for v in results.values()])
plt.ylabel("Accuracy (%)")
plt.title(f"Model Comparison")
plt.xticks(rotation=30, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()

# === Save and Report the Best Model ===
best_model_name = max(results, key=results.get)
print(f"Best Model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save the best model - local host
# if best_model is not None:
#     # Save using the SavedModel format
#     best_model.save(f"best_model_{best_model_name.replace(' ', '_')}", save_format='tf')  # Basic CNN -> Basic_CNN
#     print(f"Best model saved as: best_model_{best_model_name.replace(' ', '_')}")

# Save the best model - T4
if best_model is not None:
    # Save using the SavedModel format
    best_model.save(f"best_model_{best_model_name.replace(' ', '_')}.keras")
    print(f"Best model saved as: best_model_{best_model_name.replace(' ', '_')}.keras")
```

```python
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, recall_score

# === Load CIFAR-10 ===
from tensorflow.keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize
x_test = x_test.astype("bfloat16") / 255.0

# === Load model ===
# model = load_model("best_model_CNN_+_Augmentation_+_RES_+_SE_+_TF") # Local host
model = load_model("best_model_CNN_+_Augmentation_+_RES_+_SE_+_TF.keras")

# === Evaluate on full test set ===
y_pred_probs = model.predict(x_test)
y_pred_classes = np.argmax(y_pred_probs, axis=1)
y_true = y_test.flatten()  # Flatten for sklearn

# === Print Accuracy, F1, Recall ===
acc = accuracy_score(y_true, y_pred_classes)
f1 = f1_score(y_true, y_pred_classes, average='weighted')
recall = recall_score(y_true, y_pred_classes, average='weighted')

print(f"Accuracy: {acc:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"Recall: {recall:.4f}")

# === Pick 6 random samples for display ===
sample_indices = np.random.choice(len(x_test), 6, replace=False)
sample_images = x_test[sample_indices]
sample_labels = y_test[sample_indices]

sample_preds = model.predict(sample_images)
predicted_classes = np.argmax(sample_preds, axis=1)

# === Class names ===
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# === Display images ===
plt.figure(figsize=(12, 6))
for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.imshow(sample_images[i])
    true_label = class_names[sample_labels[i][0]]
    pred_label = class_names[predicted_classes[i]]
    plt.title(f"True: {true_label}\nPred: {pred_label}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
